```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

## Train and Test 

```python
x=iris.iloc[:,:-1]
y=iris.iloc[:,4]
```

```python
x
```

```python
y
```

## Train-Test Split 

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=4)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## Training and fitting the model

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear')
svclassifier.fit(X_train, y_train)
```

## Prediction

```python
y_pred = svclassifier.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

## Model Evaluation

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred) * 100
print("Accuracy of Model: ",accuracy)
```

```python
pd.crosstab(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

## Visualization

```python
import seaborn as sns
Iris = sns.load_dataset("Iris")
print(Iris.head())

X = Iris.drop('species',axis=1)
Y = Iris.species
```

```python
df=Iris[(Iris['species']!='virginica')]
df=df.drop(['sepal_length','sepal_width'], axis=1)
df.head()
```

```python
#convert categorical values to numerical 

df=df.replace('setosa', 0)
df=df.replace('versicolor', 1)
X=df.iloc[:,0:2]
Y=df['species']
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=Y, s=50,cmap='autumn')
```

```python
from sklearn.svm import SVC 
model = SVC(kernel='linear', C=1E10)
model.fit(X, Y)
```

```python
model.support_vectors_
```

```python
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=Y, s=50, cmap='autumn')
plt.scatter(model.support_vectors_[:,0],model.support_vectors_[:,1])
```

```python
ax = plt.gca()
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=Y, s=50, cmap='autumn')
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = model.decision_function(xy).reshape(XX.shape)

ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])

ax.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
plt.show()
```

```python
red_sample=df.sample(frac=0.7)
X=red_sample.iloc[:,0:2]
Y=red_sample['species']
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=Y, s=50, cmap='autumn')
```

## Linear Kernel

```python
from sklearn.datasets import load_iris

iris = load_iris()

a = iris.data[:,[2,3]]
b = iris.target
```

```python
plt.scatter(iris.data[:,2], iris.data[:,3],c=iris.target)
```

```python
from sklearn.svm import SVC

svc = SVC(kernel='linear')
svc.fit(iris.data[:,[2,3]], iris.target)
```

```python
h = .02
x_min, x_max = a[:, 0].min() - .5, a[:, 0].max() + .5
y_min, y_max = a[:, 1].min() - .5, a[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

```python
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
Z = Z.reshape(xx.shape)
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)
plt.scatter(a[:,0],a[:,1],c=b,s=10)
plt.scatter(svc.support_vectors_[:,0], svc.support_vectors_[:,1], c='r',s=20)
```
